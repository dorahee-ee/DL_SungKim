```python
import numpy as np
import pandas as pd
import sklearn
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
import tensorflow.compat.v1 as tf  # tensorlow 1.0으로 낮춰서 사용 => tensorflow 2.0은 session 지원 X
tf.disable_v2_behavior()
```

### 1. Build graph using TensorFlow operations

$H(x) = Wx + b$

```python
x_train = [1, 2, 3]
y_train = [1, 2, 3]

W = tf.Variable(tf.random_normal([1]), name='weight')  # Variable -> 기존의 변수와는 다른 개념 = tensorflow를 실행시키면 tensorflow가 자체적으로 변경시키는 값이다. (trainable v)
b = tf.Variable(tf.random_normal([1]), name='bias')

hypothesis = x_train * W + b
```

$cost(W, b) = -\frac{1}{m}\sum (H(x^i) - y^i)^2$

```python
cost = tf.reduce_mean(tf.square(hypothesis - y_train))  # reduce_mean = 어떤 tnesor가 주어졌을 때, 평균을 내주는 함수
```

##### Gradient Descent (cost 값을 minimize하는 방법)

```python
# minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
```

### 2. feed data and run graph (operation) = sess.run(op.feed_dict={x:x_data})
### 3. update variables in the graph (and return values)

```python
# Launch the graph in a session
sess = tf.Session()

# initializes global variables in the graph
sess.run(tf.global_variables_initializer())
```

```python
# Fit the line
for step in range(2001):
    sess.run(train)
    if step % 20 == 0:  # 20번에 한번씩 출력
        print(step, sess.run(cost), sess.run(W), sess.run(b))
```

##### - placeholders를 사용하여 코드 작성해보기

```python
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)
```

```python
for step in range(2001):
    cost_val, W_val, b_val, _ = sess.run([cost, W, b, train], feed_dict={X: [1, 2, 3], Y: [1, 2, 3]})
    if step % 20 == 0:
        print(step, cost_val, W_val, b_val)
```

placeholders를 이용하는 가장 큰 이유 중 하나는 만들어진 모델에 대해서 X, Y의 값을 따로 넘겨줄 수 있기 때문이다.
